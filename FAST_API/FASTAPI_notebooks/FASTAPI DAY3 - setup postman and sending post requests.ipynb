## GET VS POST requests

### pic1 : in the left side it is the front end of the device be it desktop or mobile. here front the device we are first sending the get request to the api which is in the right side. after getting the requests the api check and sends back the response as a desired data.

### pic2 : in the left side as usual the device with front end e.g. social media website. from the front end it sends a social media post with title and content as a data to the api. the api check and sends back the requested data back to the front end.

### difference : "get" method only sends the request to the api and takes the data as response. but the "post" method can also sends data to the api and takes response as data.
